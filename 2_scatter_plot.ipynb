```python
# import the library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# plt.scatter simple function

x = np.random.randn(100)
y = np.random.randn(100)    
plt.scatter(x, y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv(r'C:\Users\Raj\Matplotlib\datasets\batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='o')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
# size
tips = sns.load_dataset('tips')


# slower
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']) # size of the marker is determined by the size column
```

```python
# scatterplot using plt.plot
# faster
plt.plot(tips['total_bill'],tips['tip'],'o')
```

Pie Chart

1)Univariate/Bivariate Analysis
2)Categorical vs numerical
3)Use case - To find contibution on a standard scale

```python
# import the library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# simple data
data = [23,45,100,20,49] # data is in percentage with total sum of all values , it is not necessarry that total sum should be 100
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv(r'C:\Users\Raj\Matplotlib\datasets\gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode,  shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)
plt.show()
```

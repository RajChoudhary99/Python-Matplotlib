```python
# import the library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# simple bar chart

children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
```

```python
# horizontal bar chart

plt.barh(colors,children,color='black')
```

```python
# Multiple Bar charts

children1 = [10,20,40,10,30]
children2 = [20,30,10,40,20]    
width = 0.35
plt.bar(np.arange(len(children1)), children1, width, label='Children 1', color='blue')
plt.bar(np.arange(len(children2)) + width, children2, width, label='Children 2', color='red')
plt.xticks(np.arange(len(children1)) + width / 2, ['A', 'B', 'C', 'D', 'E'])
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical') # rotate the x-axis labels(for bigger names)
```

```python
# Stacked Bar chart(add the values of each bar on top of each other)

df = pd.read_csv(r'C:\Users\Raj\Matplotlib\datasets\batsman_season_record.csv')
df

plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```
